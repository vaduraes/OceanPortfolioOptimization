```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon


ShapeFileCoast="../Data_Wind/GEO_data/ne_10m_coastline.shp"
ShapeFileStates="../Data_Wind/GEO_data/ne_10m_admin_1_states_provinces_lines.shp"

df = gpd.read_file(ShapeFileCoast)
df1 = gpd.read_file(ShapeFileStates)

fig, ax = plt.subplots()
fig.set_size_inches(18.5, 10.5)

df.plot(color='black',linewidth=1,ax=ax)
df1.plot(color='black',linewidth=1,ax=ax)

ShpFilePath_BOEM="./BOEM_Renewable_Energy_Shapefiles_3/"
BOEM_ShpDir={"Wind Lease Areas":"BOEM_Wind_Leases_2_2023_carto_upgrade.shp",
            "Wind Planning Areas": "BOEM_Wind_Planning_Areas_8_2023_update.shp"}

CurrentSpeedDataPath="./CurrentData/CurrentSpeedMABSAB2009_2014_50m.npz"
CurrentSpeed_NPZ=np.load(CurrentSpeedDataPath)
CurrentSpeed=np.average(CurrentSpeed_NPZ["CurrentSpeed"],axis=1)
LatLong=CurrentSpeed_NPZ["LatLong"]


WindDataPath="../Data_Wind/WindEnergyNREL_161.0m_ATB_18MW_2030.npz"
WindSpeed_NPZ=np.load(WindDataPath)
WindCF=WindSpeed_NPZ["WindEnergy"]
WindLatLong=WindSpeed_NPZ["LatLong"]


kings_county_map = gpd.read_file(ShpFilePath_BOEM+BOEM_ShpDir["Wind Lease Areas"])
kings_county_map.plot(ax=ax, color='green', edgecolor='none', alpha=0.5,legend=True)

kings_county_map = gpd.read_file(ShpFilePath_BOEM+BOEM_ShpDir["Wind Planning Areas"])
kings_county_map.plot(ax=ax, color='red', edgecolor='none', alpha=0.5)


ax.arrow( -73.7, 37.3, 0.5,-0.5, fc="k", ec="k",head_width=0.05, head_length=0.1)
plt.text(-73.7+0.5-0.4, 37.3-0.5-0.25, 'Wind Planning Area', fontsize = 10)

ax.arrow( -73.5, 39.4, 0.5,-0.5, fc="k", ec="k",head_width=0.05, head_length=0.1)
plt.text(-73.5+0.5-0.6, 39.4-0.5-0.25, 'Wind Lease Area', fontsize = 10)

plt.scatter(LatLong[:,1],LatLong[:,0],c=CurrentSpeed,cmap="jet",s=1,alpha=0.3)
clb=plt.colorbar()

plt.scatter(WindLatLong[:,1],WindLatLong[:,0],s=1,c="k",alpha=0.3)

clb.ax.set_title('OC Speed \n[m/s]')
plt.xlim(-83,-72)
plt.ylim(32,40)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
#plt.title("MABSAB 2009-2014 Avg. OC Speed 50m and BOEM Wind Lease/Planning Areas")
```
